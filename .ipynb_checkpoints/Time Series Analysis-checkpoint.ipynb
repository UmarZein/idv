### 1. Check Missing Value

```python
# Check for missing values and their distribution
missing_values = data.isnull().sum()
missing_percentage = (missing_values / len(data)) * 100

# Create a DataFrame to display the results
missing_data_summary = pd.DataFrame({
    "Column": data.columns,
    "Missing Values": missing_values,
    "Missing Percentage": missing_percentage
}).reset_index(drop=True)
missing_data_summary
```

### 2. Convert to Date Time

```python
# Convert the 'created_at' column to datetime format
data['Time Stamp'] = pd.to_datetime(data['Time Stamp'])

# Set the 'created_at' column as the index
data.set_index('Time Stamp', inplace=True)

# Display the first few rows to confirm the changes
data.head()
```

```python
# Cek Frekuensi pengambilan data
# Calculate the time differences (frequency) between consecutive data points
time_differences = data.index.to_series().diff()

# Get the mode (most common interval) to determine the frequency
common_frequency = time_differences.mode()[0]

# Tampilkan dalam frequency summary
frequency_summary = {
    "Most Common Interval": common_frequency,
    "Unique Intervals": time_differences.unique()
}

frequency_summary
```

### 3. Plot the Data

```python
# Plot each air quality column in separate subplots
# Define a color map with enough distinct colors
color_map = plt.cm.tab10

# Select specific columns (field2 to field7) for plotting
air_quality_columns = ['MQ7','MQ135', 'Temperature', 'eCO2', 'TVOC', 'Dust Density']

for i, column in enumerate(air_quality_columns):
#for column in air_quality_columns:
    plt.figure(figsize=(6,3))
    plt.plot(data.index, data[column], label=column, color=color_map(i % 10))
    plt.title(f"Time Series Trend for {column}")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.xticks(rotation=90)
    #plt.legend(title="Metric", loc='upper right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### 4. Hitung Statistika Deskriptif

```python
# Calculate descriptive statistics: mean, median, variance
descriptive_stats = data[air_quality_columns].agg(['mean', 'median','std']).transpose()

# Rename the index to match the columns for clarity
descriptive_stats.index.name = "Column"

print("Descriptive Statistics for Air Quality Data :")
descriptive_stats
```

### 5. Cek Normality

```python
from scipy.stats import kstest, norm
from scipy.stats import shapiro

target=data['Dust Density']
stat, p = kstest(target, 'norm')
print(f"Statistic={stat}, p-value={p}")

print("Kolmogorov-Smirnov Test : ")
if p > 0.05:
    print("Data berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
print('')
print("Shapiro-Wijk Test : ")
stat, p = shapiro(target)
print(f"Statistic={stat}, p-value={p}")
if p > 0.05:
    print("Data berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

### 6. Cek Stationarity

```python
# Reloading the dataset
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller


# Select a target column for stationarity analysis
target_column = "MQ7"

# Perform Augmented Dickey-Fuller (ADF) test
adf_result = adfuller(data[target_column].dropna())

# Extract ADF test results
adf_statistic = adf_result[0]
p_value = adf_result[1]
critical_values = adf_result[4]

# Create a transposed summary DataFrame
adf_summary = pd.DataFrame({
    "Statistic": [adf_statistic],
    "p-value": [p_value],
    **{f"Critical Value {key}": [value] for key, value in critical_values.items()}
}).T

# Visualize the target column
plt.figure(figsize=(6, 3))
plt.plot(data[target_column], label=target_column, color='blue')
plt.title(f"Time Series Plot for {target_column}")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
adf_summary
```

### 7. Auto Korelasi

```python
from statsmodels.graphics.tsaplots import plot_acf

# Plot Autocorrelation Function (ACF) for each selected column
for column in air_quality_columns:
    plt.figure(figsize=(10, 3))
    plot_acf(data[column].dropna(), lags=50, title=f"Autocorrelation for {column}")
    plt.show()
```

### 8. Time Series Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform time series decomposition for each column
for column in air_quality_columns:
    plt.figure(figsize=(12, 8))
    decomposition = seasonal_decompose(data[column].dropna(), model='additive', period=24)
    
    # Plot the decomposition components
    decomposition.plot()
    plt.suptitle(f"Time Series Decomposition for {column}", fontsize=12)
    plt.tight_layout()
    plt.show()
```

# II. Analysis Time Series

### 1. Deteksi Anomaly dengan ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

# Define a function to fit ARIMA and detect anomalies
def detect_anomalies_arima(series, order=(1, 1, 1)):
    # Fit ARIMA model
    model = ARIMA(series.dropna(), order=order)
    model_fit = model.fit()
    
    # Get predictions
    predictions = model_fit.predict(start=0, end=len(series)-1, dynamic=False)
    
    # Calculate residuals
    residuals = series - predictions
    
    # Define anomaly threshold (e.g., 3 standard deviations from the mean)
    threshold = 3 * np.std(residuals.dropna())
    anomalies = residuals[np.abs(residuals) > threshold]
    
    return residuals, anomalies

# Analyze anomalies for each selected column
anomaly_results = {}
for column in air_quality_columns:
    residuals, anomalies = detect_anomalies_arima(data[column], order=(1, 1, 1))
    anomaly_results[column] = anomalies
    
    # Plot residuals and anomalies
    plt.figure(figsize=(7, 3))
    plt.plot(residuals, label="Residuals", color="blue")
    plt.scatter(anomalies.index, anomalies, color="red", label="Anomalies")
    plt.title(f"ARIMA Residuals and Anomalies for {column}")
    plt.xlabel("Time")
    plt.ylabel("Residuals")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

print("ARIMA Anomaly Detection Results :")
print({col: list(anoms.index) for col, anoms in anomaly_results.items()})
```

### 2. Dengan Random Forest

```python
data[air_quality_columns]
```

```python
# Reload the dataset
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt


# 1. Fungsi untuk mengekstraksi fitur time series data
def create_features(df, target_column, lags=[1, 2, 3], rolling_window=5):
    features = pd.DataFrame(index=df.index)
    
    # Add lag features
    for lag in lags:
        features[f"lag_{lag}"] = df[target_column].shift(lag)
    
    # Add rolling mean and rolling std
    features["rolling_mean"] = df[target_column].shift(1).rolling(rolling_window).mean()
    features["rolling_std"] = df[target_column].shift(1).rolling(rolling_window).std()
    
    # Target
    features["target"] = df[target_column]
    
    # Drop missing values
    features.dropna(inplace=True)
    return features

# 2. Gunakan Random Forest untuk anomaly detection pada 'Kolom'
target_column = "TVOC"
features = create_features(data, target_column)

# 3. Split data ,emjadi input (X) and target (y)
X = features.drop(columns=["target"])
y = features["target"]

# 4. Split kedalam train dan test sets
train_size = int(0.8 * len(features))
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

# 5. Training menggunakan Random Forest model, parameter dapat diset sesuai kebutuhan
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict and calculate residuals
y_pred = rf.predict(X_test)
residuals = y_test - y_pred

# Set threshold untuk anomaly detection, gunakan 3 kali stdev
threshold = 3 * np.std(residuals)
anomalies = residuals[np.abs(residuals) > threshold]

# Plot residual dan anomali
plt.figure(figsize=(7, 3))
plt.plot(residuals, label="Residuals", color="blue")
plt.scatter(anomalies.index, anomalies, color="red", label="Anomalies")
plt.axhline(y=threshold, color="green", linestyle="--", label="Threshold")
plt.axhline(y=-threshold, color="green", linestyle="--")
plt.title(f"Random Forest Residuals and Anomalies for {target_column}")
plt.xlabel("Time")
plt.ylabel("Residuals")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("rf.png", dpi=200, bbox_inches="tight")
plt.show()
```

### 3. Deteksi Anomaly dengan Deep Learning

```python
import os
import tensorflow as tf

## Matikan GPU jika tidak support atau tidak terdeteksi

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
print("Using TensorFlow version:", tf.__version__)
print("Num GPUs Available:", len(tf.config.experimental.list_physical_devices('GPU')))

import tensorflow as tf
print("TensorFlow version:", tf.__version__)
print("Devices:", tf.config.list_physical_devices())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Input
from tensorflow.keras.optimizers import Adam


# Preprocess the data: Normalization and scaling
def preprocess_data(data, target_column, sequence_length=50):
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data[[target_column]])
    
    X, y, timestamps = [], [], []
    for i in range(sequence_length, len(scaled_data)):
        X.append(scaled_data[i-sequence_length:i, 0])
        y.append(scaled_data[i, 0])
        timestamps.append(data.index[i])  # Store timestamps
    return np.array(X), np.array(y), scaler, timestamps

# Build GRU model
def build_gru_model(sequence_length):
    model = Sequential([
        Input(shape=(sequence_length, 1)),
        GRU(64, activation='tanh', return_sequences=True),
        GRU(64, activation='tanh', return_sequences=False),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

# Parameters
sequence_length = 50
target_column = "MQ7"

# Preprocess data
X, y, scaler, timestamps = preprocess_data(data[air_quality_columns], target_column, sequence_length)

# Split into train and test sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
timestamps_train, timestamps_test = timestamps[:train_size], timestamps[train_size:]

# Reshape data for GRU input
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build and train GRU model
gru_model = build_gru_model(sequence_length)
history = gru_model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Predict on test set
y_pred = gru_model.predict(X_test)
y_pred_rescaled = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate residuals and detect anomalies
residuals = y_test_rescaled - y_pred_rescaled
threshold = 3 * np.std(residuals)
anomalies = residuals[np.abs(residuals) > threshold]

# Plot results
plt.figure(figsize=(7, 3))
plt.plot(timestamps_test, residuals, label="Residuals", color="blue")
plt.scatter([timestamps_test[i] for i in np.where(np.abs(residuals) > threshold)[0]],
            anomalies, color="red", label="Anomalies")
plt.axhline(y=threshold, color="green", linestyle="--", label="Threshold")
plt.axhline(y=-threshold, color="green", linestyle="--")
plt.title(f"GRU Residuals and Anomalies for {target_column}")
plt.xlabel("Time")
plt.ylabel("Residuals")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("gru.png", dpi=200, bbox_inches="tight")
plt.show()
```

```python
# timestamp
idx = baseline.index

# Compute differences between consecutive entries
diffs = np.diff(idx)

# Identify the minimum time difference
min_delta = min(diffs)

# Find positions where the minimum time difference occurs consecutively
mask = (diffs[:-1] == min_delta) & (diffs[1:] == min_delta)
positions = np.where(mask)[0]

if len(positions) > 0:
    # Select a subset of three consecutive entries with uniform intervals
    pos = positions[0]
    subset = idx[pos:pos + 3]
    inferred_freq = pd.infer_freq(subset)
    print(inferred_freq)  # Output: '15T'
else:
    print("No uniform frequency found.")
```

```python
import numpy as np
from scipy import signal
sig = np.repeat([1.,1.,1.], 100)
win = signal.windows.hann(50)
filtered = signal.convolve(sig, win, mode='same') / sum(win)
```

```python
sig.shape,win.shape,filtered.shape
```

```python
fig, (ax_orig, ax_win, ax_filt) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original pulse')
ax_orig.margins(0, 0.1)
ax_win.plot(win)
ax_win.set_title('Filter impulse response')
ax_win.margins(0, 0.1)
ax_filt.plot(filtered)
ax_filt.set_title('Filtered signal')
ax_filt.margins(0, 0.1)
fig.tight_layout()
fig.show()
```
